### Build TCR-Epitope Association Matrix

```python
import pandas as pd
import numpy as np
import os

script_dir = os.path.dirname(os.path.abspath("1b-association.ipynb"))
```

```python
df_pair = pd.read_csv(os.path.join(script_dir, "../data", "tcr_epitope_pairs.tsv"), sep='\t')

data_path = os.path.join(script_dir, "../data", "unique_epitopes.tsv")
unique_epitopes = pd.read_csv(data_path, sep='\t')

data_path = os.path.join(script_dir, "../data", "unique_TCRs.tsv")
unique_tcrs = pd.read_csv(data_path, sep='\t')

association_mtx = np.zeros((len(unique_tcrs), len(unique_epitopes)), dtype=int)
```

```python
for i in range(len(df_pair)):
    tcr = df_pair.iloc[i]["TCR_alpha_CDR3"] + "-" + df_pair.iloc[i]["TCR_beta_CDR3"]
    epitope = df_pair.iloc[i]["Epitope"]

    tcr_index = unique_tcrs[(unique_tcrs["TCR_alpha_CDR3"] + "-" + unique_tcrs["TCR_beta_CDR3"]) == tcr].index[0]
    epitope_index = unique_epitopes[unique_epitopes["Epitope"] == epitope].index[0]
    
    association_mtx[tcr_index, epitope_index] = 1
```

```python
from pathlib import Path

out_dir = Path("../data")
out_dir.mkdir(exist_ok=True)
np.save(out_dir / "association.npy", association_mtx)
```
